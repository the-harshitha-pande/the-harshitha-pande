```python
# without builtin functions:
import numpy as np

# Sample data
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 4, 5])
n = len(x)

# Step 1: Means
mean_x = np.mean(x)
mean_y = np.mean(y)
print("Mean of x:", mean_x)
print("Mean of y:", mean_y)

# Step 2: Slope (m) using mean formula
num = np.sum((x - mean_x) * (y - mean_y))
den = np.sum((x - mean_x)**2)
m = num / den

# Step 3: Intercept (c)
c = mean_y - m * mean_x
print("Slope (m):", m)
print("Intercept (c):", c)

# Step 4: Predicted values (y_hat)
y_hat = m * x + c
print("Predicted values (y_hat):", y_hat)

# Step 5: Errors
errors = y - y_hat
print("Errors (y - y_hat):", errors)

# Step 6: Metrics
MAE = np.mean(np.abs(errors))
MSE = np.mean(errors**2)
RMSE = np.sqrt(MSE)
R2 = 1 - (np.sum(errors**2) / np.sum((y - mean_y)**2))

print("MAE:", MAE)
print("MSE:", MSE)
print("RMSE:", RMSE)
print("R^2 Score:", R2)
```

```python
#with builtin functions:
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Sample data
X = np.array([[1], [2], [3], [4], [5]])   # independent variable
y = np.array([2, 4, 5, 4, 5])             # dependent variable

# Step 1: Means
mean_x = np.mean(X)
mean_y = np.mean(y)
print("Mean of x:", mean_x)
print("Mean of y:", mean_y)

# Step 2: Fit model
model = LinearRegression()
model.fit(X, y)

# Step 3: Coefficients
m = model.coef_[0]
c = model.intercept_
print("Slope (m):", m)
print("Intercept (c):", c)

# Step 4: Predictions
y_hat = model.predict(X)
print("Predicted values (y_hat):", y_hat)

# Step 5: Errors
errors = y - y_hat
print("Errors (y - y_hat):", errors)

# Step 6: Metrics
MAE = mean_absolute_error(y, y_hat)
MSE = mean_squared_error(y, y_hat)
RMSE = np.sqrt(MSE)
R2 = r2_score(y, y_hat)

print("MAE:", MAE)
print("MSE:", MSE)
print("RMSE:", RMSE)
print("R^2 Score:", R2)
```

```python
#comparision table
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Sample data
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 4, 5])
X = x.reshape(-1, 1)

# ---------- Manual Calculation ----------
mean_x = np.mean(x)
mean_y = np.mean(y)

num = np.sum((x - mean_x) * (y - mean_y))
den = np.sum((x - mean_x)**2)
m_manual = num / den
c_manual = mean_y - m_manual * mean_x
y_hat_manual = m_manual * x + c_manual
errors_manual = y - y_hat_manual

MAE_manual = np.mean(np.abs(errors_manual))
MSE_manual = np.mean(errors_manual**2)
RMSE_manual = np.sqrt(MSE_manual)
R2_manual = 1 - (np.sum(errors_manual**2) / np.sum((y - mean_y)**2))

# ---------- Built-in Calculation ----------
model = LinearRegression()
model.fit(X, y)
m_builtin = model.coef_[0]
c_builtin = model.intercept_
y_hat_builtin = model.predict(X)
errors_builtin = y - y_hat_builtin

MAE_builtin = mean_absolute_error(y, y_hat_builtin)
MSE_builtin = mean_squared_error(y, y_hat_builtin)
RMSE_builtin = np.sqrt(MSE_builtin)
R2_builtin = r2_score(y, y_hat_builtin)

# ---------- Comparison Table ----------
comparison = pd.DataFrame({
    "Metric": ["Slope (m)", "Intercept (c)", "MAE", "MSE", "RMSE", "R^2 Score"],
    "Manual Calculation": [m_manual, c_manual, MAE_manual, MSE_manual, RMSE_manual, R2_manual],
    "Built-in Calculation": [m_builtin, c_builtin, MAE_builtin, MSE_builtin, RMSE_builtin, R2_builtin]
})

print(comparison)
```

```python
#📝 Example: Loading a CSV Dataset
#Built‑in (with pandas + scikit‑learn)
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import matplotlib.pyplot as plt

# Step 1: Load dataset
df = pd.read_csv("data.csv")   # replace with your file name

# Step 2: Select columns
x = df[["x"]].values   # independent variable
y = df["y"].values      # dependent variable

# Step 3: Fit regression
model = LinearRegression()
model.fit(x, y)
y_hat = model.predict(x)

# Step 4: Metrics
MAE = mean_absolute_error(y, y_hat)
MSE = mean_squared_error(y, y_hat)
RMSE = np.sqrt(MSE)
R2 = r2_score(y, y_hat)

print("Slope:", model.coef_[0])
print("Intercept:", model.intercept_)
print("MAE:", MAE, "MSE:", MSE, "RMSE:", RMSE, "R²:", R2)

# Step 5: Plot
plt.scatter(x, y, color="blue")
plt.plot(x, y_hat, color="red")
plt.show()
```
